## Create Uploader

```python
class Uploader:
    
    def __init__(self):
        file_contents = ""
    
    def upload(self):
        _upload_widget = fileupload.FileUploadWidget()
        
        def _cb(change):
            decoded = io.StringIO(change['owner'].data.decode('utf-8'))
            filename = change['owner'].filename
            print('Uploaded `{}` ({:.2f} kB)'.format(
                filename, len(decoded.read()) / 2 **10))
            self.file_contents = decoded.getvalue()
        
        _upload_widget.observe(_cb, names='data')
        display(_upload_widget)
```

```python
uploader = Uploader()
uploader.upload()
```

## Implement WordCloud

```python
class WordCloud:
    def __init__(self, file_contents, punctuations = None, uninteresting_words = None):
        self.punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
        self.uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
            "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
            "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
            "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
            "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just"
                                   ]
        self.file_contents = file_contents
        self.frequencies = {}
        
    def calculate_frequencies(self):
        # Delete all puntuations
        cleaned_text = ""
        for char in self.file_contents:
            if char not in self.punctuations:
                cleaned_text += char
        
        # Clean text from uninteresting_words which defined by user.
        words = cleaned_text.split()
        interesting_words = [word for word in words if word not in self.uninteresting_words]
        
        # Count frequency of each words
        for word in interesting_words:
            if word not in self.frequencies:
                self.frequencies[word] = 1
            else:
                self.frequencies[word] += 1
                
        cloud = wordcloud.WordCloud()
        cloud.generate_from_frequencies(self.frequencies)
        return cloud.to_array()
                
    def visualize(self):
        myimage = self.calculate_frequencies()
        plt.imshow(myimage, interpolation = 'nearest')
        plt.axis('off')
        plt.show()
```

## Test our Word Cloud

```python
word_cloud = WordCloud(file_contents = uploader.file_contents)
word_cloud.visualize()
```
